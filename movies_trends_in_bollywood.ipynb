```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("bollywood_movies.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
genre_count=dataset['Genre'].value_counts()
print("Genre Count:",genre_count)
```

```python
box_office_by_genre=dataset.groupby('Genre')['Box Office (Crores)'].sum()
print("Box Office Collection by Genre:")
print(box_office_by_genre)
```

```python
lead_actor_by_box_office=dataset.groupby('Lead Actor')['Box Office (Crores)'].sum()
print("Lead Actor with Highest Box Office Collection:")
print(lead_actor_by_box_office)
```

```python
#Bar chart of movies count by Genre
plt.figure(figsize=(15,5))
sns.countplot(x='Genre',data=dataset,palette='viridis')
plt.title("Movies count by Genre")
plt.show()
```

```python
#Box office collection by Gerne
plt.pie(box_office_by_genre,labels=box_office_by_genre.index,autopct='%0.1f%%')
plt.title("Box Office Collection by Genre")
plt.show()
```
